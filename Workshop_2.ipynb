Write a Python function that takes a number as a parameter and checks whether the number is prime or not.


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
is_prime(5)
```

Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def print_twin_primes(limit):
    for i in range(3, limit, 2):
        if is_prime(i) and is_prime(i + 2):
            print(f"({i}, {i+2})")

print_twin_primes(1000)
```

A positive integer is called a perfect number if it is equal to the sum of all of its divisors including 1 but excluding the number itself. Write a function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors.

```python
def sumOfDivisors(n):
    sum = 0
    for i in range(1,n):
        if n % i == 0:
            sum += i
    return sum

sumOfDivisors(10)
```

Write a function that converts a decimal number to binary number.

```python
def convertDeciToBinary(n):
    temp = n
    binary = 0
    while temp!=0:
        a = temp % 2
        binary = binary*10 + a
        temp = temp // 2
    return binary
convertDeciToBinary(6)
```

Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(n):
    pro= 1
    temp = n
    while temp!=0:
        a = temp % 10
        pro *= a
        temp = temp // 10
    return pro
prodDigits(234)
```
