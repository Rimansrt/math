```python
def sumDigits(x):
    total = 0
    while x != 0:
        s = x % 10
        total += s
        x = x // 10
    print("Sum of digits:", total)
sumDigits(234)
```

```python
def mypow(x,n):
  result = 1
  i = 0
  while i<n:
    result = result * x
    i=i+1
  return result
print(mypow(2,5))
```

```python
def fib():
    n = int(input("Enter the number of terms: "))
    a = 0
    b = 1
    i = 0
    print("Fibonacci sequence:")
    while i<n:
        print(a)
        c = a + b
        a = b
        b = c
        i=i+1
fib()
```

```python
hrs = input("Enter Hours: ")
h = float(hrs)

rate = input("Enter the rate per hour: ")
r = float(rate)

if h <= 40:
    pay = h * r
else:
    overtime = h - 40
    pay = (40 * r) + (overtime * r * 1.5)

print("Pay:", pay)
```

```python
def number_group(number):
  if number > 0:
    return "Positive"
  elif number == 0:
    return "Zero"
  else:
    return "Negative"
print(number_group(10))
print(number_group(0))
print(number_group(-5))
```
